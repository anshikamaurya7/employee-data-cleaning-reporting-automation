```python
names = [
    np.random.choice(first_names) + " " + np.random.choice(last_names)
    for _ in range(num_employees)
]
```

```python
gender = np.random.choice(
    ["male","Female","Male","female","M","F"],
    size=num_employees
)
```

```python
age = np.random.randint(
    21,
    60,
    
    num_employees
)
```

```python
departments = np.random.choice(
    [
        "it",
        "HR",
        "finance",
        "Sales",
        "Marketing"
    ],
    size=num_employees
)
```

```python
designations = np.random.choice(
    [
        "Intern",
        "EXecutive",
        "Analyst",
        "senior Analyst",
        "manager"
    ],
    size=num_employees
)
```

```python
experience = np.random.randint(
    0,
    20,
    num_employees
)
```

```python
salary = np.random.randint(
    25000,
    150001,
    num_employees
)
```

```python
work_mode = np.random.choice(
    [
        "OFFICE",
        "Hybrid",
        "remote"
    ],
    size=num_employees
)
```

```python
overtime = np.random.randint(
    0,
    31,
    num_employees
)
```
