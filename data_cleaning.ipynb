```python
df["Name"].unique()
```

```python
df["Name"] = df["Name"].replace({
    "Sneha mehta" : "Sneha Mehta",
    "aditya yadav": "Aditya Yadav",
    "priya Patel":"Priya Patel"})

df["Name"] = df["Name"].replace({
    "aditya Maurya": "Aditya Maurya",
    "priya Verma": "Priya Verma",
    "Neha mehta ": "Neha Mehta"})

df["Name"] = df["Name"].replace({
    "aarav Verma ": "Aarav Verma",
    "aditya Jain": "Aditya Jain",
    "Neha yadav": "Neha Yadav"
    })
```

```python
df.head()
```

```python
df["Performance_Rating"].unique()
```

```python
df["Performance_Rating"] = df["Performance_Rating"].replace({
    "GOOD" : "Good",
    "average": "Average",
    "excellent":"Excellent",
    "Poor":"Poor"
    })
```

```python
df.head()
```

```python
df["Gender"].unique()
```

```python
    
df["Gender"] = df["Gender"].replace({
    "Male" : "M",
    "male": "M",
    "female":"F",
    "Female":"F"
    })
```

```python
df.head()
```

```python
df["Department"].unique()
```

```python
df["Department"] = df["Department"].replace({
    "it" : "IT",
    "finance": "Finance",
    })
```

```python
df.head()
```

```python
df["Designation"].unique()
```

```python
df["Designation"] = df["Designation"].replace({
    "senior Analyst" : "Senior Analyst",
    "manager": "Manager",
    "EXecutive":"Executive"
    })
```

```python
df.head()
```

```python
df["City"].unique()
```

```python
df["City"] = df["City"].replace({
    "chennai" : "Chennai",
    "lucknow": "Lucknow",
    "HyderaBad":"Hyderabad",
    "DELHI":"Delhi"

    })
```

```python
df.head()
```

```python
df["Work_Mode"].unique()
```

```python
df["Work_Mode"] = df["Work_Mode"].replace({
    "OFFICE" : "Office",
    "remote": "Remote",
    "HyderaBad":"Hyderabad",
    "DELHI":"Delhi"
})
```

```python
df.head()
```

```python
df.to_csv("data/cleaned_employee_data.csv", index=False)
```
